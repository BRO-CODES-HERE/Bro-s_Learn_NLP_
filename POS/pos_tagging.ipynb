```python
import spacy 
```

```python
nlp=spacy.load("en_core_web_sm")
```

```python
with open("hello.txt", "r") as f:
    text=f.read()
```

```python
text
```

```python
doc=nlp(text)
```

```python
numerical_token=[]
noun_token=[]
```

```python
for token in doc:
    if token.pos_ == "NOUN":
        noun_token.append(token)
    elif token.pos_ == 'NUM':
        numerical_token.append(token)
```

```python
count = doc.count_by(spacy.attrs.POS)
count
```
