```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme(style="whitegrid")
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python

df.info()
display(df.describe())
display(df.describe(include=['object', 'category']))
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.title('Missing Data Map (Yellow = Missing)')
plt.show()

print("Exact Missing Counts:\n", df.isnull().sum())
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.countplot(data=df, x='survived', ax=axes[0], palette='pastel')
axes[0].set_title('Survival Count (0=No, 1=Yes)')

sns.countplot(data=df, x='pclass', ax=axes[1], palette='pastel')
axes[1].set_title('Passenger Class Distribution')

sns.countplot(data=df, x='sex', ax=axes[2], palette='pastel')
axes[2].set_title('Gender Distribution')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(df['age'].dropna(), kde=True, bins=30, ax=axes[0], color='skyblue')
axes[0].set_title('Age Distribution')

sns.histplot(df['fare'], kde=True, bins=40, ax=axes[1], color='salmon')
axes[1].set_title('Fare Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='survived', hue='sex', palette='Set2')
plt.title('Survival Breakdown by Gender')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='pclass', y='age', palette='Set3')
plt.title('Age Distribution Segmented by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
numeric_cols = df.select_dtypes(include=[np.number])
corr_matrix = numeric_cols.corr()


sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
